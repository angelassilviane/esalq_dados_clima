```python
# MÉTODO DO INTERVALO INTERQUARTIL OU IQR
# Mede a diferença entre o 1º e o 3º quartil em um determinado conjunto de dados.
# 1º quartil = 25% (q25)
# 3º quartil = 75% (q75)
# IQR = q75 - q25

# LS = q75 + (IQR * 1.5)
# LI = q25 - (IQR * 1.5)

# Para casos extremos:
# LS = q75 + (IQR * 3)
# LI = q25 - (IQR * 3)

def IQR_find_extreme_values(df, col, distancia):
    print('MÉTODO DO INTERVALO INTERQUARTIL OU IQR')
    
    # Calcular os quartis:
    Q3 = df[col].quantile(0.75)
    Q1 = df[col].quantile(0.25)

    # Calcular o IQR:
    IQR = Q3-Q1

    # Calcular o LI e LS:
    LimSup = Q3 + (IQR * distancia)
    LimInf = Q1 - (IQR * distancia)
    
    # Calcular a quantidade de valores que estão abaixo ou acima do LI e LS (valores extremos):
    df_aux = df[col]

    LS = df_aux > LimSup
    LI = df_aux < LimInf

    # Contar a quantidade de valores extremos:
    sum_LI = df_aux[LI].count()
    sum_LS = df_aux[LS].count()
    
    # Calcular o percentual:
    per_sum_LS = (sum_LS*100)/237783
    per_sum_LI = (sum_LI*100)/237783
    
    print('-----------------------------------------')
    print(col)
    print('3º quartil:',Q3)
    print('1º quartil:',Q1)
    print('LS de',col,':',LimSup)
    print('LI de',col,':',LimInf)
    print('Qtd acima de LS:',sum_LS)
    print('Qtd abaixo de LI:',sum_LI)
    print('LS(%):',per_sum_LS)
    print('LI(%):',per_sum_LI)
    print('-----------------------------------------')
```
